```python
N = 200
b = 20
ms = 4
al = .5
cols = [tableau20[0], tableau20[4]]

pts_A = np.random.multivariate_normal([-8,0], [[10,0], [0,30]], size=N)
pts_B = np.random.multivariate_normal([8,0], [[10,0], [0,30]], size=N)
pts_C = np.random.multivariate_normal([-6,6], [[20, 10], [10, 20]], size=N)
pts_D = np.random.multivariate_normal([6,-6], [[20, 5], [5, 20]], size=N)

fig, axi = plt.subplots(1, 2, figsize=(5.3,3), sharey=True)

ax = axi[0]
ax.plot(pts_A[:,0], pts_A[:,1], 'o', color=cols[0], markersize=ms, alpha=al)
ax.plot(pts_B[:,0], pts_B[:,1], 'x', color=cols[1], markersize=ms, alpha=al, markeredgewidth=1)
ax.plot([0,0], [-100,100], '-', color='black')
ax.set_xlim(-b,b)
ax.set_ylim(-b,b)
ax.set_xticks([])
ax.set_yticks([])
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_title('1D')

ax = axi[1]
ax.plot(pts_C[:,0], pts_C[:,1], 'o', color=cols[0], markersize=ms, alpha=al)
ax.plot(pts_D[:,0], pts_D[:,1], 'x', color=cols[1], markersize=ms, alpha=al, markeredgewidth=1)
ax.plot([-100,100], [-100,100], '-', color='black')
ax.set_xlim(-b,b)
ax.set_ylim(-b,b)
ax.set_xticks([])
ax.set_xlabel('Feature 1')
ax.set_title('2D')

for ax in axi:
    for i in ax.spines.itervalues():
        i.set_linewidth(0.3)


plt.tight_layout()
plt.show()
#fig.savefig(FIGDEST+'category_distributions.pdf', bbox_inches='tight')
```
